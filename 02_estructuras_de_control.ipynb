# Estructuras de control
Las estructuras de control, permiten dentro de una función controlar la ejecución de las asignaciones.
En una funcion que solo utiliza asignaciones, el control esta dado por el orden en que se escriben las asignaciones, sin embargo en algunos casos se requiere escoger la ejecucion de un grupo de asignaciones y de otro dependiendo de una condicion (estructuras de control de selección), y en otro casos se requiere repetir un conjunto de asignaciones (estructuras de control repetitivas)



se usa una estructura de selección cuando no se necesita ejecutar todas las asignaciones y por lo tanto se deben seleccionar cual ejecutar dependiendo de una condicion definida por la misma estructura de control de selección y por eso tambien se les conoce como estructuras de control condicionales



Para comprender la estructura se hace necesario tener claro lo siguiente





```[python]
if (condición) :
      instrucciones caso verdadero
      instrucciones caso verdadero
      instrucciones caso verdadero
 else:
      instrucciones caso falso
      instrucciones caso falso
      instrucciones caso falso
Instrucciones posteriores por fuera de la estructura de selección

```

En una estructura de control de seleccón se require la adecuada formulación de expresiones booleanas (condiciones) que al evalursen (calcularse) se obtiene un valor de verdad (Verdadero o falso) (True o False) y dependiendo de esto se ejecutaran las asignaciones del lado del Si(verdadero) o del lado del No (falso) de la estructura

Es así como una expresion booleana (condición) es una adecuada combinación de operandos, operadores y signos de agrupación que al ser evaluada(calculada) da como resultado un valor de verdad



La expresiones boolenas requieren la comprensión adicional a los operadores arimeticos: los operadores relacionales que permiten comparar cantidades y los operadores logicos que permiten construir condiciones compuestas. (que se cumplan todas and... o que solo se cumpla por lo menos una or)



Cada lenguaje define los simbolos para cada uno de estos operadores tanto para los relacionales como para los lógicos... en el caso de python estos son



````python
def valor_absoluto(x:float)->float:
    rta:float
    if (x<0):
       rta =-x
    else:
       rta =x
    return (rta)
````

````python
#prueba de la funcion
x:float
x=valor_absoluto(-23.2)
print("el valor absoluto dio: "+str(x))
````

````python
def mayor_de_tres (x:float,y:float,z:float)->float:
    rta:float
    rta=x
    if (y>rta):
        rta=y
    if z>rta:
        rta=z
    return (rta)
````

````python
#prueba de la función
x:float
x=mayor_de_tres(2,3,4)
print ("el mayor es: "+str(x))
````

## Estructuras de control repetitivas Mientras que

En algunas funciones se requiere repetir un grupo de asignaciones, cierto numero de veces, para lo cual se usan las estructuras de control repetitivas.




Los siguientes aspectos se deben tener en cuenta al usar una estructura de control repetitiva mientras que



El concepto de estructura de control repetitiva es independiente del lenguaje de programación y en muchos casos de aplicación se requiere contar cada vez que se repiten las asignaciones, a este proceso se le conoce como un ciclo



En python
```[python]

while (condición):
    Instrucciones a repetir
    Instrucciones a repetir
    Instrucciones a repetir
    
Instrucciones posteriores por fuera de la estructura 

```

````python
#Ejemplo para probar la estructura... calcular la sumatoria de 1 hasta n
def sumar_hasta(n:int)->int:
    s:int
    i:int
    s=0
    i=1
    while(i<=n):
        s=s+i
        i=i+1
    return (s)
````

````python
#probar la función
x:int
x=sumar_hasta(10)
print("el resultado dio: "+str(x))
````

````python
#Ejemplo de una funcón que usa una estructura repetitiva 
# funcion que cuenta los digitos de un numero entero dado
def contar_digitos (n:int)->int:
    c:int
    c=1
    while n>=10:
      c = c+1
      n = n // 10      #operador de división entera //
    return (c)
````

````python
#probar la función
x:int
x=contar_digitos(1234)
print("el resultado dio: "+str(x))
````

# Ejercicios desarrollados


````python
# dado un numero n retorna el factorial de n
def factorial (n:int)->int:
    rta:int
    i:int
    rta=1
    i=1
    while(i<=n):
        rta=rta*i
        i=i+1
    return (rta)
````

````python
#código de prueba de las funciones
x:int
x=factorial(10)
print("el resultado dio: "+str(x))
````

````python
# dado un numero entero como parametro retorna la suma de los digitos de dicho número
def sumar_digitos (num:int)->int:
    rta:int
    rta=0
    while (num>0):
        rta= rta + (num % 10)
        num= num//10
    return rta
````

````python
#código de prueba de las funciones
x:int
x=sumar_digitos(1111)
print("el resultado dio: "+str(x))
````
