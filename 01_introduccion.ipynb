## Datos
-----------
Un dato es una representación (numérica, alfabética,booleana) de un atributo o variable cuantitativa o cualitativa. Los datos son valores almacenados en la memoria de datos de un computador, y para lo cual debemos imaginarlos como casilleros ubicados en alguna dirección de memoria y el los cuales se puede guardar un único valor ya sea numerico, alfabetico o booleno



Para evitar tener que manejar los números exadecimales que representan direcciones de memoria dentro de nuestro codigo, se asigna a cada casilla contenedora de un dato un nombre identificador de la variable y un tipo de datos que representa el tipo de valor a guardar en dicha variable.
Los nombres de las variables "identificadores" solo deben contener letras del alfabeto inglés,numeros y subrayados y siempre el primer caracter debe ser una letra.



Cada variable en la memoria de datos adecuadamente identificada tiene asociado un tipo de valor que puede almacenar, y en un instante de tiempo tan solo almacenará uno y solo uno de los valores que puede contener.



Los tipos de datos deben imaginarse como conjuntos de todos los posibles valores que puede contener una variable de dicho tipo, y aunque en la vida real estos conjuntos son infinitos por ejemplo los numeros enteros, para los leguajes de programación deben ser considerados finitos, es decir que pese a que se pueden almacenar numeros muy grandes o muy pequeños, siempre hay un limite en el almacenamiento de ciertos tipos de datos.

Es asi como cuando se define una variable de un determinado tipo, significa que en dicho espacio de memoria solo se puede guardar un valor de los pertenecientes al conjunto de datos que representa dicho tipo.

## Asignación
--------------
Un espacio de memoria se denomina variable, porque su contenido puede cambiar en cualquier instante del programa, y para ello se requiere usar la instrucción mas importante en la programación que es la asignación.

Una asignación por lo tanto es el mecanismo mediante el cual se puede cambiar el valor a una variable(espacio de memoria), por el resultado de calcular una expresión

En una asignacion se deben tener en cuenta tres componentes: La variable que desea cambiar de valor, el simbolo de asignacion utilizado por el lenguaje y la expresion que se desea calcular para almacenar su valor en la variable motivo de la asignación.



En la mayoria de ejercicios de este curso, gran parte de las asignaciones requeridas se realizan sobre variables numéricas, por lo tanto las expresiones involucradas en dichas asignaciones son expresiones aritméticas. 

Una expresión aritmética se una adecuada combinación de operandos, operadores y signos de agrupación, que al ser calculada dará como resultado un valor numérico



Los simbolos utilizados para los operadores aritméticos, dependen de cada lenguaje de programación, sin embargo las operaciones basicas como suma, resta, multiplicación y división se han estandarizado en la programación de computadores. Para el caso particular de python estos son los símbolos para operadores aritmeticos:




Es importante recordar que en una asignación solo cambiará el valor de la variable de la izquierda, por lo tanto las variables involucradas en la expresion a calcular quedarán con el mismo valor, pues solo se estan utilizando estos valores para realizar los calculos que representa dicha expresión



````python
"""
Ejemplos de asignaciones
"""
#definicion de variables,  cada una asociadas a un tipo de datos
nota_1:float
nota_2:float
nota_3:float
definitiva:float
    
#asignaciones
nota_1=4.5
nota_2=3.7
nota_3=4.9
definitiva=(nota_1 + nota_2 + nota_3)/3
print("Despúes del proceso de asignación....")
print("la variable nota_1 contiene:"+str(nota_1))
print("la variable nota_3 contiene:"+str(nota_2))
print("la variable nota_3 contiene:"+str(nota_3))
print("la variable definitiva contiene:"+str(definitiva))
````

## Funciones
----------------
Una función es la unidad básica de procesamiento de datos en la programación estructurada, es decir que todo programa en el paradigma estructurado debe estar compuesto de funciones que procesan datos recibidos como parametros (entradas a la función), los procesan mediante asignaciones y estructuras de control, y produciendo un único resultado (valor de retorno).



El codigo de una función se almacena en la memoria de instrucciones, y mientras no se invoque la ejecución de la función no existen datos asociados a la función en la memoria de datos. Por lo tanto los datos utilizados en una funcion solo existirán mientras la función este en ejecución, y una vez retorne el valor correpondiente, este memoria se liberará y por lo tanto no se perderán dichos datos. 

Por lo tanto los datos de una función solo puede accederse desde la misma función que esta en ejecución.
Los datos que la funcion requiere como entrada para su procesamiento se llaman parámetros, para el procesamiento se utilizan asignaciones, las cuales normalmente requieren la definición de variables denominadas locales, y que al final de la ejecución de la función se retornará el **valor** de tan solo una de dichas variables.

En python las funciones tiene la siguiente estructura

```[python]
def nombre_funcion( parametros ):
    instrucciones de procesamiento de datos
    instrucciones de procesamiento de datos
    return (rta)
    
def otra_funcion( parametros):
    instrucciones de procesamiento
    instrucciones de procesamiento
    return (rta)
```



````python
"""
********
suma
********
suma dos números decimales (reales) cualquiera
Recibe como parámetros:
   un número decimal x cualquiera que representa el primer sumando
   un número decimal y cualquiera que representa el segundo sumando
Retorna un numero decimal que representa la suma de x mas y
Ejemplos:
    suma(2.1,4.3) retorna 6.4
    suma(-2,4) retorna 2.0
    suma(0,-4.7) retorna -4.7 
"""
def suma(x:float,y:float)->float:
    z:float
    z=x+y
    return (z)

"""
********
resta
********
resta dos números decimales (reales) cualquiera
Recibe como parámetros:
   un número decimal x cualquiera que representa el minuendo de la operación
   un número decimal y cualquiera que representa el sustraendo de la resta
Retorna un numero decimal que representa la diferencia (resultado de la resta) entre x y
Ejemplos:
    resta(2.1,4.3) retorna 2.2
    resta (-2,4) retorna-6
    resta(0,-4.7) retorna 4.7 
"""
def resta(x:float,y:float)->float:
    z:float
    z=x-y
    return (z)
````

````python
#ejemplo para probar la funcion
a:float
b:float
c:float
a=2.1
b=4.3
c=suma(a,b)
print ("el resultado de la suma es:",str(c))
c=suma(b,a)
print ("el resultado de la suma es:",str(c))
c=resta(a,b)
print ("el resultado de la resta es:",str(c))
c=resta(b,a)
print ("el resultado de la resta es:",str(c))
````
